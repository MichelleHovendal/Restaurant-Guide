```python
df = pd.read_csv('yelp_reviews_RV_categories.csv')
#df = pd.read_json('yelp_academic_dataset_business.json', lines=True)
```

```python
df = df.drop_duplicates(subset='text')
df.shape
```

```python
N = 1674264
MaxUsed = 100
df.shape
#(5177322, 23)
df = df.iloc[0:N]
df
```

```python
#Length of review
df['reviewlength'] = df['text'].str.len()
```

```python
#antal sætninger
df['sentences'] = df["text"].str.count('\.')
```

```python
#antal tal

# procentdel af store bogstaver

#procentdel af positive/negative meningsbærende ord --> Måske denne tages ud. Eller der laves liste med positive ord og en med negative


#gennemsnit længde af review
#one hot
```

```python
from gensim.parsing.preprocessing import remove_stopwords
df.text.apply(remove_stopwords) 
```

```python
from gensim.utils import simple_preprocess
# Tokenize the text column to get the new column 'tokenized_text'
df['tokenized_text'] = [simple_preprocess(line, deacc=True) for line in df['text']] 
```

```python
from gensim.parsing.porter import PorterStemmer
porter_stemmer = PorterStemmer()
# Get the stemmed_tokens
df['stemmed_tokens'] = [[porter_stemmer.stem(word) for word in tokens] for tokens in df['tokenized_text']]
```

```python
df['stemmed_text'] = df['stemmed_tokens'].apply(' '.join)
```

```python
Word = []
WordCounter = Counter(" ".join(df["stemmed_text"]).split()).most_common(MaxUsed)
for j in range(len(WordCounter)):
    Word.insert(j,WordCounter[j][0])
```

```python
Words = np.zeros((N,MaxUsed))


#for loop over rækker, og for loop over ordene i NWords. Hvis ord er der, sæt 1.
for index, row in df.iterrows():
    for i in range(np.size(Words, axis=1)):
        if Word[i] in df['stemmed_text'].iloc[index]:
            Words[index,i] = 1
```

```python
#drop attributes with text
df = df.drop(['text','tokenized_text', 'stemmed_tokens',
       'stemmed_text'], axis=1)
```

```python
df_words = pd.DataFrame(Words)

Names = list(range(0, MaxUsed))
ColNames = [str(n) for n in Names]

df_words.columns = ColNames
df_words
```

```python
df = df.join(df_words)

df
```

```python
df['user_id'] = df['user_id'].astype('category')
df['business_id'] = df['business_id'].astype('category')

df['UserCat'] = df['user_id'].cat.codes
df['BusinessCat'] = df['business_id'].cat.codes

df
```

```python
#f_feature = df.drop[['name','address', 'city', 'state', 'postal_code','latitude', 'longitude','attributes','categories','hours','cat_kitchen', 'cat_type','text', 'tokenized_text', 'stemmed_tokenz', 'stemmed_text']]
#df_feature = df[['stars', 'review_count','review_stars', 'reviewlength', 'sentences', '1','2','3','4','5','6']]

df_feature = df.drop(['business_id','name', 'address','city', 'state', 'postal_code', 'latitude',
       'longitude', 'attributes', 'categories',
       'hours', 'cat_kitchen', 'cat_type', 'user_id','date'], axis=1)
df_feature
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
model=IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.2),max_features=1.0)
model.fit(df_feature)
```

```python
df['scores']=model.decision_function(df_feature)
df['anomaly']=model.predict(df_feature)
```
